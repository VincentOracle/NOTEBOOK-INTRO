```python

#KMeans Clustering
# Machine learning applications of unsupervised learning include K-means clustering. 
# To divide a dataset into K clusters, where K is a predetermined number of clusters, 
# is the objective of k-means clustering
```

```python

#Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Loading the Wine Quality dataset
wine_data = pd.read_csv("C:\\Users\\n\\Downloads\\WineQT.csv")
wine_data.head()
```

```python
wine_data.tail()
```

```python
# Extract features to be used for clustering
X = wine_data.drop([ 'quality', 'Id'], axis=1)

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

```

```python
# Fit K-Means clustering algorithm with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_std)

# Assign clusters to each data point
wine_data['Cluster'] = kmeans.labels_

# Print cluster centers
print(kmeans.cluster_centers_)
```

```python
#Expectation Maximization:

# A strategy for carrying out maximum likelihood estimation in
# the presence of latent variables is the expectation-maximization algorithm.
# To achieve this, the model is first optimized, then the latent variable values are estimated, 
# and so on until convergence.
```

```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler


# Extract features to be used for clustering
X = wine_data.drop([ 'quality', 'Id'], axis=1)

# Standardize the data
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Fit Expectation Maximization algorithm with 3 clusters
em = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
em.fit(X_std)

# Assign clusters to each data point
wine_data['Cluster'] = em.predict(X_std)

# Print cluster means
print(em.means_)
```

```python
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.decomposition import PCA
import pandas as pd

# Load data
data = pd.read_csv("C:\\Users\\n\\Downloads\\WineQT.csv")
```

```python
#importing the required libaries
from sklearn.decomposition import FastICA
import pandas as pd

# Load data
data = pd.read_csv("C:\\Users\\n\\Downloads\\WineQT.csv")
# Separate features and labels
X = data.drop(['quality', 'Id'], axis=1)
y = data['quality']

# Apply ICA
ica = FastICA(n_components=5)
X_ica = ica.fit_transform(X)

# Save ICA results to new CSV file
result = pd.DataFrame(X_ica, columns=['IC1', 'IC2', 'IC3', 'IC4', 'IC5'])
result['quality'] = y
result['sulphates'] = data['sulphates']
result.to_csv('Wine_Quality_Data_ICA.csv', index=False)
print(result)
```

```python
#Importing the neccessary libraries
from sklearn.random_projection import GaussianRandomProjection
import pandas as pd
```

```python
# Load data

data = pd.read_csv("C:\\Users\\n\\Downloads\\WineQT.csv")
# Separate features and labels
X = data.drop(['quality', 'chlorides', 'Id'], axis=1)
y = data['quality']

# Apply Randomized Projections
rp = GaussianRandomProjection(n_components=5)
X_rp = rp.fit_transform(X)

# Save Randomized Projections results to new CSV file
result = pd.DataFrame(X_rp, columns=['RP1', 'RP2', 'RP3', 'RP4', 'RP5'])
result['quality'] = y
result['chlorides'] = data['chlorides']
result.to_csv('Wine_Quality_Data_RP.csv', index=False)
print(result)
```
